Create a Jupyter notebook, import matplotlib. Write cells that create an array x ranging from [0,1] in 100 steps and that defines functions that return sin(x) and cos(x).  In a new cell use to create a multipanel plot (1 row, 2 columns), plotting sin(x) vs. x in the left panel and cos(x) vs. x in the right panel. Label the panels with sin(x) and cos(x), and save the figure as a PDF.


```python
import matplotlib.pyplot as plt
import numpy as np
import math
import sys
```

```python

#creates array x
x = np.linspace(0,1,100)

#defines functions for sin and cos
def sin_function(x):
    return np.sin(x)


def cos_function(x):
    return np.cos(x)

#creates plot figure

plt.figure(figsize=(12,6))

# sin vs x plot
plt.subplot(1,2,1)
plt.plot(x, sin_function(x))
plt.title("sin(x) vs. x")
plt.xlabel("x")
plt.ylabel("sin(x)")


# cos vs x plot
plt.subplot(1,2,2)
plt.plot(x, cos_function(x))
plt.title("cos(x) vs. x")
plt.xlabel("x")
plt.ylabel("cos(x)")





plt.savefig("sin_cos_plot.pdf")



```
